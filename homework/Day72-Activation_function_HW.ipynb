# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    return x * (x > 0)

def dReLU(x):
    return (x > 0)
```

```python
x = plt.linspace(-10,10,100)

plt.plot(x, ReLU(x), c='r', label='linspace(-10,10,100)')
plt.title('ReLU Function')

plt.show()
```

```python
plt.plot(x, dReLU(x), c='b', label='linspace(-10,10,100)')
plt.title('dReLU Function')

plt.show()
```
